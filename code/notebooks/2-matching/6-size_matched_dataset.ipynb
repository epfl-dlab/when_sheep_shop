```python
import pandas as pd
import numpy as np

data_folder = '../data/'
```

```python
df = pd.read_csv(data_folder + 'matched/breweries.csv')
print(len(df))
```

```python
df = pd.read_csv(data_folder + 'matched/beers.csv', header=[0,1])

print()

for min_nbr in [0, 5, 10, 20]:
    subdf = df[(df['ba']['nbr_ratings'] >= min_nbr) & (df['rb']['nbr_ratings'] >= min_nbr)]
    
    print('Min {} ratings:'.format(min_nbr))
    print('  Number of breweries: {}'.format(len(subdf['ba']['brewery_id'].unique())))
    print('  Number of matched beers: {}'.format(len(subdf)))
    print('  Nbr ratings BA: {}'.format(np.sum(subdf['ba']['nbr_ratings'])))
    print('  Nbr ratings RB: {}'.format(np.sum(subdf['rb']['nbr_ratings'])))
    print()
```
